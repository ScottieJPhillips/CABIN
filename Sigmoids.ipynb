```python
import torch
import matplotlib.pyplot as plt

SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
x=torch.linspace(-4,4,101)
```

```python
sigmoid=torch.sigmoid(x)
sigmoid02=torch.sigmoid(2*x)
sigmoid15=torch.sigmoid(15*x)
stepfunction=torch.heaviside(x,torch.tensor([0.0]))
```

```python

plt.plot(x,stepfunction,label="$\\Theta(x)$")
plt.plot(x,sigmoid15,label="$\\sigma(15x)$")
plt.plot(x,sigmoid02,label="$\\sigma(2x)$")
plt.plot(x,sigmoid,label="$\\sigma(x)$")
plt.legend(loc="lower right")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.savefig("figures/sigmoids.pdf",bbox_inches='tight')
```
