```python
import json
jsonpath = '/data/ealmazan/WPs/silo_medium_mc23a/'
jsonfile = 'medium_smootheff_rad_Z.json'
json_file_full = f'{jsonpath}{jsonfile}'
with open(json_file_full, 'r') as file:
    data = json.load(file)

def extract_ranges(data):
    pt_ranges = []
    eta_ranges = []
    for category, pt_data in data.items():
        for pt_range in pt_data.keys():
            pt_start, pt_end = map(float, pt_range.split('_')[1:])
            pt_ranges.append((pt_start, pt_end))
        for pt_range, eta_data in pt_data.items():
            for eta_range in eta_data.keys():
                eta_start, eta_end = map(float, eta_range.split('_')[1:])
                eta_ranges.append((eta_start, eta_end))
    pt_ranges_sorted = sorted(set(pt_ranges), key=lambda x: x[0])
    eta_ranges_sorted = sorted(set(eta_ranges), key=lambda x: x[0])
    return pt_ranges_sorted, eta_ranges_sorted
pt_ranges_sorted, eta_ranges_sorted = extract_ranges(data)

print('PT RANGES')
print(pt_ranges_sorted)

print('ETA RANGES')
print(eta_ranges_sorted)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import torch

def data_mask(pt_ranges, eta_ranges, d):
    mask = (d['y_pt_custom'] >= pt_ranges[0]) & (d['y_pt_custom'] < pt_ranges[1]) & \
           (abs(d['y_eta_custom']) >= eta_ranges[0]) & (abs(d['y_eta_custom']) < eta_ranges[1])
    return d[mask]
print(len(pt_ranges_sorted))
print(len(eta_ranges_sorted))
def x_y_splits(pt_ranges, eta_ranges, sgn, bkg, features):
    x_train = [[None for _ in range(len(pt_ranges))] for _ in range(len(eta_ranges))]
    x_test  = [[None for _ in range(len(pt_ranges))] for _ in range(len(eta_ranges))]
    y_train = [[None for _ in range(len(pt_ranges))] for _ in range(len(eta_ranges))]
    y_test  = [[None for _ in range(len(pt_ranges))] for _ in range(len(eta_ranges))]
    sc_list = [[StandardScaler() for _ in range(len(pt_ranges))] for _ in range(len(eta_ranges))]

    for i, eta_range in enumerate(eta_ranges):         
        for j, pt_range in enumerate(pt_ranges):    
            sgn_data = data_mask(pt_range, eta_range, sgn)
            bkg_data = data_mask(pt_range, eta_range, bkg)
            
            print(f'for pt = {pt_range} and eta = {eta_range}')
            print(f'SGN:{len(sgn_data)}, BKG:{len(bkg_data)}')
            if len(sgn_data) ==0 and len(bkg_data) == 0:
                continue
            X_sgn = np.column_stack([sgn_data[f].to_numpy() for f in features]) if len(sgn_data) > 0 else np.empty((0, len(features)))
            X_bkg = np.column_stack([bkg_data[f].to_numpy() for f in features]) if len(bkg_data) > 0 else np.empty((0, len(features)))
            y_sgn = np.ones(len(X_sgn))
            y_bkg = np.zeros(len(X_bkg))

            X = np.concatenate([X_sgn, X_bkg], axis=0)
            y = np.concatenate([y_sgn, y_bkg], axis=0)

            stratify_y = y if len(np.unique(y)) > 1 else None

            X_train_split, X_test_split, Y_train_split, Y_test_split = train_test_split(
                X, y, test_size=0.3, random_state=42, stratify=stratify_y
            )

            scaler = sc_list[i][j]
            X_train_scaled = scaler.fit_transform(X_train_split)
            X_test_scaled = scaler.transform(X_test_split)

            x_train[i][j] = torch.tensor(X_train_scaled, dtype=torch.float32).detach()
            x_test[i][j]  = torch.tensor(X_test_scaled, dtype=torch.float32).detach()
            y_train[i][j] = torch.tensor(Y_train_split, dtype=torch.float32).detach()
            y_test[i][j]  = torch.tensor(Y_test_split, dtype=torch.float32).detach()
            # print(len(x_test[i][j]),len(y_test[i][j]))
    return x_train, x_test, y_train, y_test, sc_list


features = ['y_Rhad','y_Reta','y_Rphi','y_weta2','y_Eratio','y_deltae','y_wtots1','y_fracs1','y_weta1']
x_train, x_test, y_train, y_test, scalers = x_y_splits(pt_ranges_sorted, eta_ranges_sorted, sgn, bkg, features)

```

```python
def getcuts(net,scalers):
    eta=net.eta
    pt = net.pt
    m=net.features
    scaled_cuts = [[[[0] for _ in range(len(pt))] for _ in range(len(eta))] for _ in range(m)]
    unscaled_cuts=[[[[0] for _ in range(len(pt))] for _ in range(len(eta))] for _ in range(m)]
    for i in range(len(eta)):
        for j in range(len(pt)):
            cuts=net.nets[i][j].get_cuts().detach().numpy()
            uncuts=scalers[i][j].inverse_transform(cuts.reshape(1, -1))[0]
            for p in range(m):
                cutval=uncuts[p]
                uncutval=uncuts[p]
                scaled_cuts[p][i][j]=cutval
                unscaled_cuts[p][i][j]=uncutval
    return scaled_cuts, unscaled_cuts

scaled_cuts,unscaled_cuts = getcuts(net,scalers)
```
